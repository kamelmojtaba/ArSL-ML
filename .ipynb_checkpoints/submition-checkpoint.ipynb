## Data Preprocessing

Loading the data

```python
dataset_path = 'out'
```

```python
batch_size = 32
img_height = 64
img_width = 64
num_classes =32
IMAGE_SIZE = (img_height, img_width)
```

```python
full_dataset = image_dataset_from_directory(
    dataset_path,
    seed=123,
    image_size=IMAGE_SIZE,
    batch_size=batch_size)
```

function to display the images

```python
def show_batch(image_batch, label_batch):
    plt.figure(figsize=(10,10))
    for n in range(32):
        ax = plt.subplot(6,6,n+1)
        plt.imshow(image_batch[n].numpy().astype("uint8"))
        plt.title(full_dataset.class_names[label_batch[n]])
        plt.axis("off")
```

Split the data 

```python
val_size = int(len(full_dataset) * 0.1)
test_size = int(len(full_dataset) * 0.1)
```

```python
full_dataset_length = tf.data.experimental.cardinality(full_dataset).numpy()
val_size = int(full_dataset_length * 0.1)
test_size = int(full_dataset_length * 0.1)

train_ds = full_dataset.skip(val_size + test_size)
val_ds = full_dataset.take(val_size)  
test_ds = full_dataset.skip(val_size).take(test_size)  
```

Configure the dataset for performance

```python
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
test_ds = test_ds.cache().prefetch(buffer_size=AUTOTUNE)

```

Normalize the pixel value

```python
normalization_layer = tf.keras.layers.experimental.preprocessing.Rescaling(1./255)

train_ds = train_ds.map(lambda x, y: (normalization_layer(x), y))
val_ds = val_ds.map(lambda x, y: (normalization_layer(x), y))
test_ds = test_ds.map(lambda x, y: (normalization_layer(x), y))
```

Display the images and labels

```python
image_batch, label_batch = next(iter(train_ds))

image_batch = image_batch * 255.0  

show_batch(image_batch, label_batch)
```

# Modeling

## ResNet101V2

ResNet101V2 is an image recognition convolutional neural network trained to distinguish objects in images. It is a subset of the ResNet101 network, which is a more in-depth version of the original ResNet network. The "V2" in the name alludes to an improved version of the original ResNet101 network.


### Build The Transfer Model

Initializing the model

```python
base_model = ResNet101V2(weights='imagenet', include_top=False, input_shape=(*IMAGE_SIZE, 3))
```

Unfreeze the top layers of the base model

```python
base_model.trainable = True
fine_tune_at = 100  
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False
```

Building the model

```python
ResNet_model = Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),
    Dense(num_classes, activation='softmax')
])
```

Compile the model

```python
ResNet_model.compile(optimizer=Adam(learning_rate=0.0001),
              loss=SparseCategoricalCrossentropy(),
              metrics=['accuracy'])
```

##### Model Summary

```python
ResNet_model.summary()
```

 Callbacks

```python
callbacks = [
    EarlyStopping(monitor='val_loss', patience=10),
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.00001)
]
```

### Training the model

```python
ResNet_history = ResNet_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    callbacks=callbacks
)

# Evaluate the model
print("Performance on test set:")
ResNet_model.evaluate(test_ds)
```

### Plot Training History

```python
ResNet_acc = ResNet_history.history['accuracy']
ResNet_val_acc = ResNet_history.history['val_accuracy']

ResNet_loss = ResNet_history.history['loss']
ResNet_val_loss = ResNet_history.history['val_loss']
```

Training and Validation Accuracy

```python
plt.figure(figsize=(10, 10))
plt.subplot(2, 1, 1)
plt.plot(ResNet_acc, label='Training Accuracy')
plt.plot(ResNet_val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.ylabel('Accuracy')
#plt.ylim([min(plt.ylim()),1])
plt.title('ResNet Training and Validation Accuracy')
plt.show()
```

Training and Validation Loss

```python
plt.figure(figsize=(10, 10))
plt.subplot(2, 1, 2)
plt.plot(ResNet_loss, label='Training Loss')
plt.plot(ResNet_val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.ylabel('Cross Entropy')
#plt.ylim([0,1.0])
plt.title('ResNet Training and Validation Loss')
plt.xlabel('epoch')
plt.show()
```

## Convolutional Neural Network(CNN)

Conventional neural networks (CNN) is a supervised type of Deep learning, mainly used for image and speech recognition. In the sequential model includes:
rescaling, 2D Convolution, 2D Max Pooling, Flattening and dense layers.

### Build The Model

Initializing the model

```python
cnn_model = Sequential([
    # First Convolutional Block
    Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(64, 64, 3)),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    Dropout(0.2),
    
    # Second Convolutional Block
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    Dropout(0.2),
    
    # Third Convolutional Block
    Conv2D(128, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    Dropout(0.3),

    # Fourth Convolutional Block
    Conv2D(256, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(2, 2),
    Dropout(0.4),
    
    # Flattening the output and adding dense layers
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(len(full_dataset.class_names), activation='softmax')  
])
```

 Compile the model

```python
cnn_model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

#### Model Summary

```python
cnn_model.summary()
```

callbacks

```python
callbacks = [
    EarlyStopping(monitor='val_loss', patience=10),
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.00001)
]
```

### Training the model

```python
epochs = 50

cnn_history = cnn_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=epochs,
    verbose=1,
    callbacks=callbacks

)
print("PERFOMANCE",cnn_model.evaluate(test_ds))
```

### Plot Training History

```python
cnn_acc = cnn_history.history['accuracy']
cnn_val_acc = cnn_history.history['val_accuracy']

cnn_loss = cnn_history.history['loss']
cnn_val_loss = cnn_history.history['val_loss']
```

Training and Validation Accuracy

```python
plt.figure(figsize=(10, 10))
plt.subplot(2, 1, 1)
plt.plot(cnn_acc, label='Training Accuracy')
plt.plot(cnn_val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.ylabel('Accuracy')
#plt.ylim([min(plt.ylim()),1])
plt.title('CNN Training and Validation Accuracy')
plt.show()
```

Training and Validation Loss

```python
plt.figure(figsize=(10, 10))
plt.subplot(2, 1, 2)
plt.plot(cnn_loss, label='Training Loss')
plt.plot(cnn_val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.ylabel('Cross Entropy')
#plt.ylim([0,1.0])
plt.title('CNN Training and Validation Loss')
plt.xlabel('epoch')
plt.show()
```

## Long Short-Term Memory (LSTM)

Long Short-Term Memory (LSTM) networks are a specialized form of recurrent neural networks (RNN) designed for tasks requiring long-term contextual understanding, such as language modeling and text generation. They effectively manage sequence data by using input, forget, and output gates to control the flow of information, allowing them to remember and utilize past context over extended sequences.

### Build Up The Model

Initializing the images shape

```python
input_shape = (img_height, img_width, 3)  # Original shape of each image
```

Initializing the model

```python

lstm_model = models.Sequential([
    # Flatten the height and channels together while keeping the width intact
    layers.Reshape((img_height, img_width * 3), input_shape=input_shape),
    # LSTM layers
    layers.LSTM(64, return_sequences=True),
    layers.LSTM(32),
    # Dropout to prevent overfitting
    layers.Dropout(0.5),
    # Dense layer for classification
    layers.Dense(64, activation='relu'),
    # Output layer
    layers.Dense(num_classes, activation='softmax')
])
```

Compile the model

```python
lstm_model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              metrics=['accuracy'])
```

### Model Summary

```python
lstm_model.summary()
```

 callbacks

```python
callbacks = [
    EarlyStopping(monitor='val_loss', patience=10),
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.00001)
]
```

### Training the model

```python
lstm_history = lstm_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=60,
    callbacks=callbacks
)

print("PERFOMANCE",lstm_model.evaluate(test_ds))
```

### Plot Training History

```python
lstm_acc = lstm_history.history['accuracy']
lstm_val_acc = lstm_history.history['val_accuracy']

lstm_loss = lstm_history.history['loss']
lstm_val_loss = lstm_history.history['val_loss']
```

Training and Validation Accuracy

```python
plt.figure(figsize=(10, 10))
plt.subplot(2, 1, 1)
plt.plot(lstm_acc, label='Training Accuracy')
plt.plot(lstm_val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.ylabel('Accuracy')
#plt.ylim([min(plt.ylim()),1])
plt.title('LSTM Training and Validation Accuracy')
plt.show()
```

Training and Validation Loss

```python
plt.figure(figsize=(10, 10))
plt.subplot(2, 1, 2)
plt.plot(lstm_loss, label='Training Loss')
plt.plot(lstm_val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.ylabel('Cross Entropy')
#plt.ylim([0,1.0])
plt.title('LSTM Training and Validation Loss')
plt.xlabel('epoch')
plt.show()
```

# Evaluation 

create a function to evluate the models 

```python
def evaluate_model(model, test_ds, average='macro'):
    
    test_labels = np.concatenate([y for x, y in test_ds], axis=0)
    predictions = model.predict(test_ds)
    predicted_labels = np.argmax(predictions, axis=1)  # Use for multi-class classification
    
    # Calculate metrics
    accuracy = accuracy_score(test_labels, predicted_labels)
    precision = precision_score(test_labels, predicted_labels, average=average)
    recall = recall_score(test_labels, predicted_labels, average=average)
    f1 = f1_score(test_labels, predicted_labels, average=average)
    
    return accuracy, precision, recall, f1
```

```python
ResNet_metrics = evaluate_model(ResNet_model, test_ds)
cnn_metrics = evaluate_model(cnn_model, test_ds)
lstm_metrics = evaluate_model(lstm_model, test_ds)
```

Print the results

```python
metrics_names = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
models_names = ['ResNet', 'CNN', 'LSTM']

for model_name, metrics in zip(models_names, [ResNet_metrics, cnn_metrics, lstm_metrics]):
    print(f"{model_name} Evaluation:")
    for metric_name, metric_value in zip(metrics_names, metrics):
        print(f"  {metric_name}: {metric_value*100:.2f}%")
    print("") 
```

### comparing the models

Prepare data for plotting

Extracting model performances for each metric

```python
resnet_performance = [ResNet_metrics[i] for i in range(len(ResNet_metrics))]
cnn_performance = [cnn_metrics[i] for i in range(len(cnn_metrics))]
lstm_performance = [lstm_metrics[i] for i in range(len(lstm_metrics))]
```

Ploting
